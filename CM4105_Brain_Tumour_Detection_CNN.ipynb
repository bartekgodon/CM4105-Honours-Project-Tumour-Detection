# CM4105 Brain Tumour Detection CNN

```python
# Import Modules
import pathlib

import tensorflow as tf

from tensorflow.keras import datasets, layers, models
from tensorflow.keras.models import Sequential

import matplotlib.pyplot as plt
import numpy as np
```

## 1. Data Pre-procesing and Exploration

### 1.1. Dataset Import

```python
# Constants (Change as necessary)
DIRECTORY = "Dataset" 
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32

# Training Set
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DIRECTORY,
    validation_split=0.2,
    subset="training",
    seed=1337,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE)

# Validation Set
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DIRECTORY,
    validation_split=0.2,
    subset="validation",
    seed=1337,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE)
```

### 1.2. Visualise Imported Data

```python
plt.figure(figsize=(10, 10))

for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i+1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(int(labels[i]))
        plt.axis("off")

plt.savefig("9x9_ds_sample.jpeg")        
```

```python
# Get the number of classes
class_names = train_ds.class_names
NUM_CLASSES = len(class_names)
```

```python
for image_batch, labels_batch in train_ds:
  print(image_batch.shape)
  print(labels_batch.shape)
  break
```

### 1.3. Normalise the Dataset

```python
# The RGB channel values are in the range of [0, 255], however, 
# the CNN requires these to be standardised in the range [0, 1]
# The normalization_layer will be used as a seperate layer in the model.

normalization_layer = layers.experimental.preprocessing.Rescaling(1./255, 
                                                                 input_shape=(224, 224, 3))
```

### 1.4. Optimise Dataset for Performance

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## 2. Experiment 1 - Base CNN Architecture

### 2.1. Build Basic Model Architecture

```python
## CNN Model (Sequential Model)
cnn = Sequential([
    normalization_layer,
    
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    
    layers.Dense(NUM_CLASSES),
    layers.Activation(activation='softmax')
])
```

### 2.2. Compile the Model

```python
# Set Compilation Parameters
LOSS = 'binary_crossentropy'
OPTIMIZER = 'adam'
METRICS = ['accuracy']

cnn.compile(optimizer=OPTIMIZER,
              loss=LOSS,
              metrics=METRICS)
```

### 2.3. Model Summary

```python
# Print the CNN Setup Summary
cnn.summary()
```

### 2.4. Train the CNN Model

```python
# Set Training Parameters
EPOCHS = 10

# Train (fit) the model
cnn_history = cnn.fit(train_ds, 
                      validation_data=val_ds,
                      epochs=EPOCHS)
```

### 2.5. Visualise the Results

```python
# Results (Accuracy and Validation Accuracy)
accuracy = cnn_history.history['accuracy']
val_accuracy = cnn_history.history['val_accuracy']

# Results (Loss and Validation Loss)
loss = cnn_history.history['loss']
val_loss = cnn_history.history['val_loss']

epochs_range = range(EPOCHS)

# Plot the results...
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, accuracy, label='Training Accuracy')
plt.plot(epochs_range, val_accuracy, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss', alpha=0.5)
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

# Show the graphs
plt.show()
plt.savefig("base_line_results.jpeg")
```

## 3. Improving Model Performance - Revised Model Architecture

### 3.1. Build the Revised Model

```python
## CNN Model (Sequential Model)
cnn_v2 = Sequential([
    normalization_layer,
    
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Dropout(0.2), # Added Dropout Layer (Reduce Overfitting)
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid') # Sigmoid output layer
])
```

### 3.2. Compile the Model

```python
# Set Compilation Parameters
LOSS = 'binary_crossentropy'
OPTIMIZER = 'adam'
METRICS = ['accuracy']

cnn_v2.compile(optimizer=OPTIMIZER,
              loss=LOSS,
              metrics=METRICS)
```

### 3.3. Train the Model

```python
# Set Training Parameters
EPOCHS = 10

# Train (fit) the model
cnn_v2_history = cnn_v2.fit(train_ds, 
                      validation_data=val_ds,
                      epochs=EPOCHS)
```

### 3.4. Visualise the Results

```python
# Results (Accuracy and Validation Accuracy)
accuracy_v2 = cnn_v2_history.history['accuracy']
val_accuracy_v2 = cnn_v2_history.history['val_accuracy']

# Results (Loss and Validation Loss)
loss_v2 = cnn_v2_history.history['loss']
val_loss_v2 = cnn_v2_history.history['val_loss']

epochs_range = range(EPOCHS)

# Plot the results...
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, accuracy_v2, label='Training Accuracy')
plt.plot(epochs_range, val_accuracy_v2, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss_v2, label='Training Loss', alpha=0.5)
plt.plot(epochs_range, val_loss_v2, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

# Show the graphs
plt.show()
plt.savefig("revised_model_results.jpeg")
```

## 4. Improving Model Performance - Data Augmentation

### 4.1. Creating the new Data Augmentation layers

```python
# Data Augmentation Layer
# Contains random horizontal, vertical and horizontal/vertical flipping

# Horizontal
aug_horizontal = Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", input_shape=(224, 224, 3), name="Horizontal_Flip"),
  ]
)

# Vertical
aug_vertical = Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("vertical", input_shape=(224, 224, 3), name="Vertical_Flip"),
  ]
)

# Horizontal and Vertical 
aug_horizontal_vertical = Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical", input_shape=(224, 224, 3), name="Hor_Ver_Flip"),
  ]
)

# Array of Augmentations
augmentations = [aug_horizontal, aug_vertical, aug_horizontal_vertical]
```

### 4.2. Visulise the Data Augmentations

```python
plt.figure(figsize=(10, 10))

for images, _ in train_ds.take(1):
    for i in range(3):
        augmented_images = aug_horizontal(images)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_images[0].numpy().astype("uint8"))
        plt.axis("off")
        
plt.savefig("augmentation_example.jpeg")
```

### 4.3. Create the new CNN Architecture

```python
def build_aug_model(aug_layer):
    cnn = Sequential([
        normalization_layer,
        aug_layer,

        layers.Conv2D(16, 3, padding='same', activation='relu'),
        layers.MaxPooling2D(),

        layers.Conv2D(32, 3, padding='same', activation='relu'),
        layers.MaxPooling2D(),

        layers.Conv2D(64, 3, padding='same', activation='relu'),
        layers.MaxPooling2D(),

        layers.Dropout(0.2), # Added Dropout Layer (Reduce Overfitting)

        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid') # Sigmoid output layer
    ])
    
    # Set Compilation Parameters
    LOSS = 'binary_crossentropy'
    OPTIMIZER = 'adam'
    METRICS = ['accuracy']

    cnn.compile(optimizer=OPTIMIZER,
                  loss=LOSS,
                  metrics=METRICS)
    
    return cnn
```

```python
EPOCHS = 10

# Lists for storing the results
aug_accuracy = []
aug_val_accuracy = []
aug_loss = []
aug_val_loss = []

for augmentation in augmentations:
    print("Training model. Augmentation Layer: " + augmentation.layers[0].name)
    aug_cnn = build_aug_model(augmentation)
    
    # Run the model
    aug_cnn_history = aug_cnn.fit(train_ds, 
                                  validation_data=val_ds,
                                  epochs=EPOCHS)
    
    # Results (Accuracy and Val Accuracy)
    accuracy = aug_cnn_history.history['accuracy']
    val_accuracy = aug_cnn_history.history['val_accuracy']

    # Results (Loss and Validation Loss)
    loss = aug_cnn_history.history['loss']
    val_loss = aug_cnn_history.history['val_loss']
    
    # Add the results
    aug_accuracy.append(accuracy)
    aug_val_accuracy.append(val_accuracy)
    aug_loss.append(loss)
    aug_val_loss.append(val_loss)
    
    print("Training Complete!")
```

```python
# Results (Accuracy and Validation Accuracy)
accuracy = aug_accuracy[0]
val_accuracy = aug_val_accuracy[0]

# Results (Loss and Validation Loss)
loss = aug_loss[0]
val_loss = aug_val_loss[0]

epochs_range = range(EPOCHS)

# Plot the results...
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, accuracy, label='Training Accuracy')
plt.plot(epochs_range, val_accuracy, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel("Epoch")
plt.ylabel("Accuracy %")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss', alpha=0.5)
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")

# Show the graphs
plt.show()
plt.savefig("horizontal_aug.jpeg")
```

## 5. Transfer Learning

### 5.1. Rescale Pixel Values

```python
# Layer for pre-processing values to [-1, 1] range as expected MobleNetV2
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
```

### 5.2. Create the Base Model

```python
# Create the base model from the pre-trained model MobileNet V2
IMAGE_SHAPE = IMAGE_SIZE + (3,)

mobile_net_v2 = tf.keras.applications.MobileNetV2(input_shape=IMAGE_SHAPE,
                                                  include_top=False,
                                                  weights='imagenet')
```

```python
# Freeze the Convolutional Base before compiling and training
# Prevents weights in a given layer from being updated
mobile_net_v2.trainable = False
```

```python
# View the model architecture
mobile_net_v2.summary()
```

```python
# Add a classification head
# Global Average Pooling 2D - layer to convert the features into a single 1280-element vector
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()

# Dense Layer - convert features into a single prediction; no activation as predictions
# are treated as a "logit" (raw prediction value)
prediction_layer = tf.keras.layers.Dense(1)
```

```python
# Build the model by chaining the above layers...
inputs = tf.keras.Input(shape=(224, 224, 3))
x = preprocess_input(inputs)
x = mobile_net_v2(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
mnv2_model = tf.keras.Model(inputs, outputs)
```

```python
# Compile the model
# Binary Crossentropy since there are two classes
# Base Learning Rate ????
# Logits=True, as model provides a linear output
base_learning_rate = 0.0001

mnv2_model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

### 5.3. Train the MobileNetV2 Model

```python
# Epochs as 10 to keep consistend with the other experiments
EPOCHS = 10

# Model fitting...
mnv2_model_history = mnv2_model.fit(train_ds, 
                                    epochs=EPOCHS,
                                    validation_data=val_ds)
```

### 5.4. Visualise the Results

```python
acc = mnv2_model_history.history['accuracy']
val_acc = mnv2_model_history.history['val_accuracy']

loss = mnv2_model_history.history['loss']
val_loss = mnv2_model_history.history['val_loss']

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.show()
```

### 5.5. MobileNetV2 with Data Augmentation

```python
# Create the model with a data augmentation layer (horizontal and vertical flip)
aug_ver_hor = augmentations[2]

# Build the model by chaining the layers and base models...
inputs = tf.keras.Input(shape=(224, 224, 3))
x = aug_ver_hor(inputs)
x = preprocess_input(inputs)
x = mobile_net_v2(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
mnv2_aug_model = tf.keras.Model(inputs, outputs)
```

```python
# Compile the model
# Binary Crossentropy since there are two classes
# Base Learning Rate ????
# Logits=True, as model provides a linear output
base_learning_rate = 0.0001

mnv2_aug_model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
                       loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                       metrics=['accuracy'])
```

### 5.6. Train the MobileNetV2 with Data Augmentation

```python
# Epochs as 10 to keep consistend with the other experiments
EPOCHS = 10

# Model fitting...
mnv2_aug_model_history = mnv2_aug_model.fit(train_ds, 
                                            epochs=EPOCHS,
                                            validation_data=val_ds)
```

### 5.7. Visualise the Results

```python
acc = mnv2_aug_model_history.history['accuracy']
val_acc = mnv2_aug_model_history.history['val_accuracy']

loss = mnv2_aug_model_history.history['loss']
val_loss = mnv2_aug_model_history.history['val_loss']

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')

plt.show()
```
